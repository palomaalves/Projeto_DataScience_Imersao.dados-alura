Desafio 04: Estudar sobre as estatíticas retornadas no .describe()

Index(['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'], 

Mean, Média aritmética - É soma de todas as medições divididas pelo número de observações no conjunto de dados.

Std, Standart Desviation ou Desvio padrão - também representado como a letra grega sigma σ é uma medida de dispersão em torno da média populacional de uma variável aleatória. Referência

50%, median ou Mediana - É uma medida de localização do centro da distribuição dos dados, definida do seguinte modo: ordenados os elementos da amostra, a mediana é o valor (pertencente ou não à amostra) que a divide ao meio, isto é, 50% dos elementos da amostra são menores ou iguais à mediana e os outros 50% são maiores ou iguais à mediana.

25%, percentile ou 1º quartil - É uma medida de localização do 1/4 dos dados, de forma similar a mediana tem como objetivo encontrar os elementos que são igual a 25% ou menor.

75% ou 3º quartil - É uma medida de localização do 3/4 dos dados, de forma similar a mediana tem como objetivo encontrar os elementos que são igual a 75% ou maior.

Min e Max, Mínimo e Máximo - São numeros absolutos encontrados o menor e maior valor, ou 0% e 100% definindo assim o intervalo total da amostra.

```python
import numpy as np

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

####Desafio 01: Criar tabelas de frequencia com o pandas.groupby()


```python
# Agregando com a contagem de composto para cada index e dividindo pelo número total de experimento;
# Multiplicando por 100 para porcentagem e arredondando para 2 casas decimais.
round(dados.groupby(by=['dose', 'tempo', 'tratamento']).count()[['composto']] / dados.shape[0] * 100, 2)
```

```python
# Aplicando o unstack() para desagregar o tratamento
dados.groupby(by=['dose', 'tempo', 'tratamento']).count()[['composto']].unstack()
```

####Desafio 02: Normalizar o crosstab pela coluna

```python
# Usando o parâmetro normalize='columns' para obter a frequencia pela coluna.
pd.crosstab(index=[dados['dose'], dados['tempo']], columns=dados['tratamento'], normalize='columns')
```

####Desafio 03: Explorar outros agregadores

```python
# Explorando em loop outros agreadores no crosstab
mask=['min', 'max', 'mean', 'std']
for func in mask:
  print('\n', func)
  print(pd.crosstab(index  = [dados.dose, dados.tempo], 
                    columns= dados.tratamento, 
                    values = dados['g0'], aggfunc=func))
```

####Desafio 04: Explorar o Melt

```python
# O pd.melt() "derrete" as colunas "value_vars" 
# pivotando ela em apenas uma coluna enquanto mantém as outras com id_vars.

columns=dados.loc[:, 'g0': 'g9'].columns
pd.melt(dados, id_vars=['dose', 'tempo', 'tratamento'], value_vars=columns)
```

```python
#Usando o melt para descobrir o composto com maiores expressoes.
pd.melt(dados, id_vars='g0', value_vars='composto').sort_values(by='g0')
```

####Desafio 05: Calcular e analisar a correlação entre G e C

```python
corr_GenCel = dados.loc[:, 'g0':'g7'].merge(dados.loc[:, 'c0':'c7'], on=dados['id']).corr()
corr_GenCel
```

```python
cmap = sns.diverging_palette(230, 20, as_cmap=True)

mask = np.zeros_like(corr_GenCel)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr_GenCel, mask=mask, cmap=cmap, center=0,
                     linewidths=.5, cbar_kws={"shrink": .5})
```
